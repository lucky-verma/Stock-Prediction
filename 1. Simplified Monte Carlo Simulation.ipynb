```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = yf.download("TSLA")
```

```python
returns = np.log(1+ df['Adj Close'].pct_change())
```

```python
mu, sigma = returns.mean(), returns.std()
```

```python
simRetruns = np.random.normal(mu, sigma, 90)
```

```python
initial = df['Adj Close'].iloc[-1]
```

```python
simPrices = initial * (simRetruns + 1).cumprod()
```

```python
plt.plot(simPrices)
```

```python
for i in range(100):
    simRetruns = np.random.normal(mu, sigma, 90)
    simPrices = initial * (simRetruns + 1).cumprod()
    # plt.axhline(initial, c='k')
    plt.plot(simPrices)
```
